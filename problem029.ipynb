# Distinct powers
## Problem 29

Consider all integer combinations of ab for $2 ≤ a ≤ 5$ and $2 ≤ b ≤ 5$:

<center>$2^2=4, 2^3=8, 2^4=16, 2^5=32$</center>
<center>$3^2=9, 3^3=27, 3^4=81, 3^5=243$</center>
<center>$4^2=16, 4^3=64, 4^4=256, 4^5=1024$</center>
<center>$5^2=25, 5^3=125, 5^4=625, 5^5=3125$</center>

If they are then placed in numerical order, with any repeats removed, we get the following sequence of 15 distinct terms:

<center>$4, 8, 9, 16, 25, 27, 32, 64, 81, 125, 243, 256, 625, 1024, 3125$</center>

How many distinct terms are in the sequence generated by $a^b$ for $2 ≤ a ≤ 100$ and $2 ≤ b ≤ 100$?

```python
a_range = range(2, 101)
b_range = range(2, 101)
```

```python
answer = len({ a ** b for a in a_range for b in b_range})
print("Answer is ", answer)
```
